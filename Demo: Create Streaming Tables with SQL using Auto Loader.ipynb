### Demo: Create Streaming Tables with SQL using Auto Loader
We will keep the same data and test Streaming Table.
This method is recommanded over COPY _INTO_

```python
%sql
-- Load our catalog and schema
USE CATALOG workspace;
USE SCHEMA learning;

-- Check current catalog and schema
SELECT current_catalog(), current_schema()
```

```python
%sql
CREATE OR REFRESH STREAMING TABLE sql_autoloader
SCHEDULE EVERY 1 WEEK
AS
SELECT *
FROM STREAM read_files(
  "dbfs:/Volumes/workspace/learning/learn/nyctaxis_stream",
  format => "parquet"
);
```

```python
%sql
SELECT *
FROM sql_autoloader
```

```python
%sql
DESCRIBE TABLE EXTENDED sql_autoloader
```

```python
%sql
REFRESH STREAMING TABLE sql_autoloader
```

```python
%sql
SELECT * 
FROM sql_autoloader
```
